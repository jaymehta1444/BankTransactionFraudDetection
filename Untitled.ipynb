We remove the outliers from the 'amount' column only as that is the only numeric data feature that directly affects the target variable.

```python
def remove_outliers_iqr(df, features):
    for feature in features:
        Q1 = df[feature].quantile(0.25)
        Q3 = df[feature].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        df = df[(df[feature] >= lower_bound) & (df[feature] <= upper_bound)]
    return df
df = remove_outliers_iqr(df, ['amount'])
df.shape
```

For the features 'nameOrig' and 'nameDest', we only consider the first letter, which is either M or C. This decides if it is a Merchant or Customer transaction.

```python
def get_first_letter(name):
    return name[0] if isinstance(name, str) and len(name) > 0 else None
df['nameOrigfirstletter'] = df['nameOrig'].apply(get_first_letter)
df['namedestfirstletter'] = df['nameDest'].apply(get_first_letter)
```
